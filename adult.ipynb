#Replace Salary Values ['<=50K','>50K'] with 0 and 1

```python
df['income'].value_counts()
```

```python
sns.countplot(data=df, x='income')
```

```python
def income(inc):
  if inc=='<=50K':
    return 0
  else:
    return 1

df['newIncome'] = df['income'].apply(income)
df.head(5)
```

```python
df.replace(to_replace=['<=50K','>50K'],value = [0,1],inplace = True)
```
